```python
# This is used to get the useful columns and unuseful columns
from sklearn.decomposition import PCA
import pandas as pd

# Assume df is your DataFrame with features
# X is your feature matrix
X = df.values

# Initialize PCA with desired number of components
pca = PCA(n_components=2)  # Example: 2 components for simplicity

# Fit PCA on your data
pca.fit(X)

# Get the explained variance ratio for each component
explained_variance_ratio = pca.explained_variance_ratio_

# Get the components (loadings) to understand how original features contribute to each component
components = pd.DataFrame(pca.components_, columns=df.columns)

# Now, you can analyze 'components' DataFrame to see how original features contribute to each component
# You can sort them based on their importance in explaining variance

# For example, you can filter out features based on a threshold of their contribution to components
threshold = 0.1  # Adjust as needed
important_features = components.abs().max() > threshold
useful_columns = df.columns[important_features]
unuseful_columns = df.columns[~important_features]

print("Useful Columns:", useful_columns)
print("Unuseful Columns:", unuseful_columns)
```
