<h2>Evaluating for Missing Data</h2>

The missing values are converted to default. We use the following functions to identify these missing values. There are two methods to detect missing data:

<ol>
    <li><b>.isnull()</b></li>
    <li><b>.notnull()</b></li>
</ol>
The output is a boolean value indicating whether the value that is passed into the argument is in fact missing data.

```python
#this is used to know total number of null values in each attribute
null_values_sum=df.isnull().sum()
null_values_sum
```

```python
#Here wherever there is a null values it will show 'True' and were there is not null it will show 'False'
df.isnull()
#And the vice versa of .isnull command is notnull
df.notnull() 
```

```python
#Now we will store these boolean values in a variable and check the total no. of missing values in a column
missing_values = df.isnull()
missing_values
```

```python
# To know the total no. of misssing values in a particular column we will run a for loop 
missing_values=df.isnull()
for column in headers:
    print(column)
    print(missing_values[column].value_counts())
    print(" ") #This is to give space between 
```

<br>
<h3>How to deal with missing data?</h3>

<ol>
    <li>drop data<br>
        a. drop the whole row<br>
        b. drop the whole column
    </li>
    <li>replace data<br>

```python
# WE calculate the avg(mean) of normalized-losses
avg_nor_loss= df['normalized-losses'].astype("float").mean() 
avg_nor_loss
```

```python
df['normalized-losses'].replace(np.nan,avg_nor_loss,inplace=True)
df.head(15)
```

```python
# This is used to count the no. of counts in the particular attribute 
df['num-of-doors'].value_counts()
df['body-style'].value_counts()
```

```python
# This is used to count the MAX no. of counts of value in the particular attribute 
df['num-of-doors'].value_counts().idxmax()
```

```python
# So we will replace null value with the max value in the column
df['num-of-doors'].replace(np.nan, 'four', inplace=True)
```

```python
df.head()
```

```python
#Command to delete the rows in specific columns having NULL Values
df.dropna(subset=['price'] ,inplace=True, axis=0)

#reset the index value of the newly formed data
df.reset_index(drop=True, inplace=True)
```

# Preprocessing using Scikit-Learn

```python
#Import api of scikit learn here
from sklearn.impute import SimpleImputer
impute_mean= SimpleImputer #this we creates a object in python using simple-imputer
```

```python
#Here this command is used to replace the null values with the mean 
impute_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
#And now we will give command to execute all this in the main dataset
#This command doesent do any changes in the original dataset as this command dont support inplace=True
impute_mean.fit_transform(df[['normalized-losses','stroke','bore','peak-rpm']]).astype('object')
df[['normalized-losses','stroke','bore','peak-rpm']]= impute_mean.fit_transform(df[['normalized-losses','stroke','bore','peak-rpm']]).astype('object')
```

```python
df.head(14)
```

```python
imputer_most_frequent= SimpleImputer(missing_values=np.nan, strategy='most_frequent')
df[['num-of-doors']]= imputer_most_frequent.fit_transform(df[['num-of-doors']])
```
